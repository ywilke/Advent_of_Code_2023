```python
from aocd.models import Puzzle
```

```python
puzzle = Puzzle(year=2023, day=1)
print(puzzle.input_data)
```

```python
input_list = puzzle.input_data.split("\n")
print(len(input_list), "input_items")

calibration_values = []
first_value = None
last_value = None
found_first = False
for line in input_list:
    for char in line:
        try:
            value = int(char)
            last_value = value
            if found_first is False:
                first_value = value
        except ValueError:
            pass
    line_value = int(f"{first_value}{last_value}")
    calibration_values.append(line_value)

calibration_sum = sum(calibration_values)
print("calibration sum is:",calibration_sum)

puzzle.answer_a = calibration_sum
```
